Preparing Data:

```
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
import itertools
from sklearn.model_selection import cross_validate
```

```
df = pd.read_csv("world_cup22.csv")

df.loc[:, "month":] = df.loc[:, "month":].apply(lambda col: StandardScaler().fit_transform(col.values.reshape(-1, 1)).flatten(), axis=0)

df
```

```
X_cols = df.loc[:, "month":]
X_cols = X_cols.columns
X_train, X_test, y_train, y_test = train_test_split(df[X_cols], df["number of goals"], train_size = 60, test_size =36, random_state= 18)
```

```
LinReg = LinearRegression()
LinReg.fit(X_train, y_train)
y_train_pred = LinReg.predict(X_train)
mean_squared_error(y_train, y_train_pred)
```

```
y_test_pred = LinReg.predict(X_test)
mean_squared_error(y_test, y_test_pred)
```

These 2 value suggest overfitting because the test error is much higher than the train error suggesting that the model is not fitting the data properly. 

```
import altair as alt

df_coef = pd.DataFrame({
    'feature': LinReg.feature_names_in_,
    'coef': LinReg.coef_
})

chart = alt.Chart(df_coef).mark_bar().encode(
    x='coef',
    y=alt.Y('feature') 
).properties(
    title='Coefficients'
)
chart
```

```
months_index = df.columns.get_loc("month")
column_names = df.columns[months_index:].tolist()
three_tuples = list(itertools.combinations(column_names, 3))
rng = np.random.default_rng(seed = 129)
random_tuples = rng.choice(three_tuples, size = 1000)
random_tuples
```

Cross Validation

```
mse_list = []

X_train_df = pd.DataFrame(X_train, columns=column_names)

for triple in random_tuples:
    indices = [column_names.index(col) for col in triple]
    
    
    X_subset = X_train_df.iloc[:, indices].values
    
    cv_results = cross_validate(LinearRegression(), X_subset, y_train, cv=10, scoring='neg_mean_squared_error')
    cv_mse = -np.mean(cv_results['test_score'])
    mse_list.append((tuple(triple), cv_mse))
mse_list
```

```
mse_dict = dict(mse_list)
mse_series = pd.Series(mse_dict)
best_triple = mse_series.idxmin()
best_triple
```

```
best_mse = mse_series.min()
best_mse
```

```
best_cols = ['completed line breaks', 'crosses', 'goal preventions']
X_train2, X_test2, y_train2, y_test2 = train_test_split(df[best_cols], df["number of goals"], train_size = 60, test_size =36, random_state= 18)
```

```
LinReg.fit(X_train2,y_train2)
y_train_pred2 = LinReg.predict(X_train2)
mean_squared_error(y_train2, y_train_pred2)
```

```
y_test_pred2 = LinReg.predict(X_test2)
mean_squared_error(y_test2, y_test_pred2)
```
