```
import pandas as pd
import altair as alt
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsRegressor
import numpy as np
from sklearn.metrics import mean_squared_error
```

1.

```
n= 2000
rng = np.random.default_rng()
x_random = rng.uniform(low = -5, high = 5, size = n)
epsilon = rng.normal( loc = 0, scale = 5, size = n)
y_random = 3*(x_random**2) + epsilon

df = pd.DataFrame({"x": x_random, "y": y_random})

alt.Chart(df).mark_point().encode(
    x = "x",
    y = "y"
)

```

2.

```
def get_error(train_size, k, set_used):
    X_train, X_test, y_train, y_test = train_test_split(df[["x"]], df["y"], train_size = train_size, random_state= 18)
    regression_model = KNeighborsRegressor(n_neighbors = k)
    regression_model.fit(X_train, y_train)
    if set_used == "train":
        y_pred = regression_model.predict(X_train)
        return  mean_squared_error(y_train, y_pred)
    if set_used == "test":
        y_pred = regression_model.predict(X_test)
        return mean_squared_error(y_test, y_pred)
    
    if set_used != "train" and set_used != "test":
        return print("Error: Please indicate set used (test/ train)")
    




```

3.

```
k_vector = np.arange(100,2,-1)
k_val = np.zeros(k_vector.size)
train_error = np.zeros(k_vector.size)
test_error = np.zeros(k_vector.size)
train_size = 1600
train_size_input = train_size/2000
for i in range(k_vector.size) :
    k_val[i] = 1/k_vector[i]
    train_error[i] = get_error(train_size_input,k_vector[i],'train')
    test_error[i] = get_error(train_size_input,k_vector[i],'test')


error_df = pd.DataFrame({"k": k_val, "train error": train_error,"test error": test_error})
```

```
train_curve = alt.Chart(error_df).mark_line(color='red').encode(
    x = "k",
    y = "train error"
    
    
)
test_curve = alt.Chart(error_df).mark_line().encode(
    x = "k",
    y = "test error"
    
)
test_curve + train_curve
```
